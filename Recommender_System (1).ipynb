# Apriori Algorithm

````python
from mlxtend.frequent_patterns import association_rules, apriori
````

````python
transactions_str = df.groupby(['BillNo', 'Itemname'])['Itemname'].count().reset_index(name ='Count')
transactions_str
````

````python
# making a mxn matrice where m=BillNo and n=itemname and each row represents whether the item was in the transaction or not
my_basket = transactions_str.pivot_table(index='BillNo', columns='Itemname', values='Count', aggfunc='sum').fillna(0)


my_basket.head()
````

````python
# making a function which returns 0 or 1
# 0 means item was not in that transaction, 1 means item present in that transaction

def encode(x):
    if x<=0:
        return 0
    if x>=1:
        return 1

# applying the function to the dataset

my_basket_sets = my_basket.applymap(encode)
my_basket_sets.head()
````

````python
# using the 'apriori algorithm' with min_support=0.03(3% of 581587)
# It means the item should be present in atleast 174 transaction out of 581587 transactions only when we considered that item in
# frequent itemset

from mlxtend.frequent_patterns import apriori,association_rules
frequent_items = apriori(my_basket_sets, min_support = 0.03,use_colnames = True)
frequent_items
````

````python
# now making the rules from frequent itemset generated above

rules = association_rules(frequent_items, metric = "lift", min_threshold = 1)
rules.sort_values('confidence', ascending = False, inplace = True)
rules
````

# **Recommender System**

````python
def rec_sys(user_itemset, dataset):
    items_list = dataset[rules['antecedents'] == user_itemset]
    items_list = items_list[items_list['confidence'] >= 0.03] 
    items_list["consequents"] = items_list["consequents"].apply(lambda x: ', '.join(list(x))).astype("unicode") 

    items_in_set = set() # to get unduplicate item we use set
    for item_sets in items_list['consequents']: 
        for item in item_sets.split(','):
            items_in_set.add(item.strip()) 
    
    print(f'You can also buy these items with: {user_itemset}')
    for items in items_in_set:
        print(items)
````
